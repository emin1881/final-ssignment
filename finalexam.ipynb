```python
import yfinance as yf
import plotly.graph_objects as go
import pandas as pd
def make_graph(stock_data, revenue_data, stock):
    fig = go.Figure()
    fig.add_trace(
        go.Scatter(
            x=stock_data.index,
            y=stock_data['Close'],
            mode='lines',
            name="Share Price",
            line=dict(color='blue'),
            yaxis="y1"
        )
    )
    fig.add_trace(
        go.Scatter(
            x=revenue_data.index,
            y=revenue_data['Revenue'],
            mode='lines',
            name="Revenue",
            line=dict(color='green'),
            yaxis="y2"
        )
    )
    fig.update_layout(
        title=stock,
        xaxis=dict(title="Date"),
        yaxis=dict(title="Price (USD)", side='left'),
        yaxis2=dict(title="Revenue (USD Millions)", side='right', overlaying='y')
    )
    fig.show()
stock_data = yf.Ticker("TSLA").history(period="max")
revenue_data = pd.DataFrame({
    'Date': pd.date_range(start='2021-01-01', periods=10, freq='M'),
    'Revenue': range(100, 110)
}).set_index('Date')
make_graph(stock_data, revenue_data, "Tesla (TSLA)")
```

```python
import yfinance as yf
import plotly.graph_objects as go
import pandas as pd

def make_graph(stock_data, revenue_data, stock):

    fig = go.Figure()


    fig.add_trace(
        go.Scatter(
            x=stock_data.index,
            y=stock_data['Close'],
            mode='lines',
            name="Share Price",
            line=dict(color='blue'),
            yaxis="y1"
        )
    )


    fig.add_trace(
        go.Scatter(
            x=revenue_data.index,
            y=revenue_data['Revenue'],
            mode='lines',
            name="Revenue",
            line=dict(color='green'),
            yaxis="y2"
        )
    )


    fig.update_layout(
        title=stock,
        xaxis=dict(title="Date"),
        yaxis=dict(title="Price ($US)", side='left'),
        yaxis2=dict(title="Revenue ($US Millions)", side='right', overlaying='y')
    )


    fig.show()


stock_data = yf.Ticker("GME").history(period="max")
revenue_data = pd.DataFrame({
    'Date': pd.date_range(start='2021-06-01', periods=10, freq='M'),
    'Revenue': range(100, 110)
}).set_index('Date')


make_graph(stock_data, revenue_data, "GameStop (GME)")
```
